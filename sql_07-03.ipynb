#INNER JOIN

```python
query='SELECT * FROM cust_details cd INNER JOIN order_details od on cd.cust_id=od.cust_id'
run=pd.read_sql(query,engine)
run
```

```python
query='''SELECT cd.cust_name,cd.cust_id,COUNT(cd.cust_id) as frequency 
FROM cust_details cd INNER JOIN order_details od on cd.cust_id=od.cust_id
GROUP BY cd.cust_id
ORDER BY frequency DESC;'''
run=pd.read_sql(query,engine)
run
```

```python
query='SELECT * FROM cust_details cd LEFT JOIN order_details od on cd.cust_id=od.cust_id'
run=pd.read_sql(query,engine)
run
```

```python
query='SELECT * FROM cust_details cd RIGHT JOIN order_details od on cd.cust_id=od.cust_id'
run=pd.read_sql(query,engine)
run
```

```python
query='SELECT * FROM cust_details cd FULL OUTER JOIN order_details od on cd.cust_id=od.cust_id'
run=pd.read_sql(query,engine)
run
```

```python
query='SELECT * FROM cust_details cd INNER JOIN order_details od on cd.cust_id=od.cust_id AND od.order_qty>2 AND od.item_id in (5,1,12,18)'
run=pd.read_sql(query,engine)
run
```
